```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("M16-Amazon-Challenge").config("spark.driver.extraClassPath","/content/postgresql-42.2.16.jar").getOrCreate()
```

```python
from pyspark import SparkFiles
url = "https://s3.amazonaws.com/amazon-reviews-pds/tsv/amazon_reviews_us_Watches_v1_00.tsv.gz"
spark.sparkContext.addFile(url)
df = spark.read.option("encoding", "UTF-8").csv(SparkFiles.get("amazon_reviews_us_Watches_v1_00.tsv.gz"), sep="\t", header=True, inferSchema=True)
df.show()
```

```python
from pyspark.sql.functions import to_date
# Read in the Review dataset as a DataFrame
watch_reviews_df = df.dropna()
watch_reviews_df.show()
```

```python
# Create the vine_table. DataFrame
vine_df = watch_reviews_df.select(["review_id","star_rating","helpful_votes","total_votes","vine","verified_purchase"])
vine_df.show()
```

```python
# Filtered the dataframe for more than 20 total vote counts.
vine_filtered_votes_df = vine_df.filter("total_votes>=20")
vine_filtered_votes_df.show()
```

```python
# Filtered the dataframe  where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
vine_helpful_total_votes_df = vine_filtered_votes_df.filter("(helpful_votes/total_votes)>=0.5")
vine_helpful_total_votes_df.show()
```

```python
# Filtered the dataframe to retrieve all the rows where a review was written as part of the Vine program (paid).
vine_paid_df = vine_helpful_total_votes_df.filter("vine == 'Y'")
vine_paid_df.show()
```

```python
# Filtered the dataframe to retrieve all the rows where a review was written as part of the Vine program (unpaid).
vine_unpaid_df = vine_helpful_total_votes_df.filter("vine == 'N'")
vine_unpaid_df.show()
```

```python
# Total number of reviews.
vine_helpful_total_votes_count = vine_helpful_total_votes_df.count()
vine_helpful_total_votes_count
```

```python
# Total number of paid reviews.
total_paid_reviews = vine_paid_df.count()
total_paid_reviews
```

```python
# Total number of unpaid reviews.
total_unpaid_reviews = vine_unpaid_df.count()
total_unpaid_reviews
```

```python
# Total number of paid 5-star reviews.
paid_five_star_reviews = vine_paid_df.filter(vine_paid_df.star_rating == 5).count()
paid_five_star_reviews
```

```python
# Total number of unpaid 5-star reviews.
unpaid_five_star_reviews = vine_unpaid_df.filter(vine_unpaid_df.star_rating == 5).count()
unpaid_five_star_reviews
```

```python
# Paid 5-star reviews percentage.
paid_five_star_reviews_percent = (paid_five_star_reviews / total_paid_reviews) * 100
paid_five_star_reviews_percent
```

```python
# Unpaid 5-star reviews percentage.
unpaid_five_star_reviews_percent = (unpaid_five_star_reviews / total_unpaid_reviews) * 100
unpaid_five_star_reviews_percent
```
